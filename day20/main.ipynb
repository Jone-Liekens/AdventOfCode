```python
# part 2 more efficient

import os, re, sys
import numpy as np

filename = "data.txt"
with open(filename, 'r') as f:
    s = f.read()
    rows = s.split("\n")

nx, ny = len(rows[0]), len(rows)
board = np.array([[{'.': 0, '#': 1, 'S': 5, 'E': 9}[x] for x in row] for row in rows], dtype=np.int8)
ys, xs = np.array(np.where(board == 5))[:, 0]
ye, xe = np.array(np.where(board == 9))[:, 0]
board[ys, xs], board[ye, xe] = 0, 0

def valid_point(point):
    return 0 <= point[0] < ny and 0 <= point[1] < nx

def find_next(point, prev_point):
    y, x = point 
    adjacent = [(y-1, x), (y+1, x), (y, x+1), (y, x-1)]
    for point in adjacent:
        if valid_point(point) and not point == prev_point:
            if board[point] == 0:
                return point

# first we try to find the normal path
y, x = ys, xs
prev_y, prev_x = 0, 0

point = y, x 
prev_point = prev_y, prev_x
path = {point: 0}
i = 0
while point != (ye, xe):
    new_point = find_next(point, prev_point)
    prev_point = point
    point = new_point

    i += 1
    path[point] = i

d = 20 # max timeout
count = 0
def search_shortcuts(y, x):
    global count, d
    current = path[y, x]
    for dy in range(-d, d + 1):
        for dx in range(-d + abs(dy), d - abs(dy) + 1):
            newy, newx = y + dy, x + dx
            if valid_point((newy, newx)):
                if board[newy, newx] == 0:
                    gain = path[newy, newx] - current - abs(newy - y) - abs(newx - x)
                    count += int(gain >= 100)

for y in range(ny):
    for x in range(nx):
        if board[y, x] == 0:
            search_shortcuts(y, x)


```
