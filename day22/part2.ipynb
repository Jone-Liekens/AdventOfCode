```python

# part 1

import os, re, sys
import numpy as np
import math

filename = "test.txt"
with open(filename, 'r') as f:
    s = f.read()
    
startcodes = [int(x) for x in s.split("\n")]
startcodes
def step1(secr):
    return ((64 * secr) ^ secr) % 16777216

def step2(secr):
    return (math.floor(secr / 32) ^ secr) % 16777216

def step3(secr):
    return ((2048 * secr) ^ secr) % 16777216


#startcodes = [123]
n = 2000

startcodes = [1, 2, 3, 2024]

prices = np.zeros((len(startcodes), n+1), dtype=int)
deltas = np.zeros((len(startcodes), n), dtype=int)



total_score = 0
for code_i in range(len(startcodes)):
    secret = startcodes[code_i]
    price = str(secret)[-1]

    prices[code_i][0] = price

    for i in range(n):
        secret = step3(step2(step1(secret)))
        price = int(str(secret)[-1])

        prices[code_i, i+1] = price
        deltas[code_i, i] = prices[code_i, i+1] - prices[code_i, i]

    total_score += secret

print(total_score)


def calculate_policy(series):
    k = len(series)
    total = 0
    for i in range(len(startcodes)):
        j = 0
        while j < n - k + 1:
            if (deltas[i, j:j+k] == series).all():
                total += prices[i, j+k]
                break
            j += 1
    return total




# we guess that the last one should be positive

ds = {}
for d in range(-9, 10):
    s = calculate_policy(np.array([d]))
    ds[d] = s


#for d in ds:
    #print(d, ds[d])



optimal_ds = sorted(ds, key=lambda x: ds[x], reverse=True)

print(optimal_ds)
print([int(ds[d]) for d in optimal_ds])


ds = {}
for d in range(-9, 10):
    s = calculate_policy(np.array([d]))
    ds[d] = s

best_ds = sorted(ds, key=lambda x: ds[x], reverse=True)
highest_possible_ds = [int(ds[d]) for d in optimal_ds]

i = 0
max_val = 0

cs = {}

while max_val < highest_possible_ds[i]:
    d = best_ds[i]

    for c in range(-9, 10):
        s = calculate_policy(np.array([c, d]))
        cs[(c, d)] = s

        if s > max_val:
            max_val = s

    i += 1

for key in cs:
    print(key, cs[key])

```
