```python
import numpy as np

filename = "data.txt"
with open(filename, 'r') as f:
    s = f.read()
    board = s.split("\n")

ny = len(board)
nx = len(board[0])

seen_positions = np.zeros((ny, nx))
current_area = []
current_borders = []


def is_valid(y, x):
    return 0 <= x < nx and 0 <= y < ny



def discover_region(y, x, name):
    adjacent = [(y+1, x), (y-1,x), (y, x+1), (y,x-1)]
    borders = [(y+0.1, x), (y-0.1,x), (y, x+0.1), (y,x-0.1)]
 
    for i in range(4):
        element = adjacent[i]
        ny, nx = element
        if is_valid(ny, nx):
            
            if board[ny][nx] == name:

                if not seen_positions[ny][nx]:
                    seen_positions[ny, nx] = 1
                    current_area.append(element)
                    discover_region(ny, nx, name)
                continue

        border = borders[i]
        if border not in current_borders:
            current_borders.append(border)




def count_borders():

    count = len(current_borders)

    for i in range(len(current_borders)):
        y, x = current_borders[i]
        if x % 1 == 0:
            # y is a half
            for j in range(0, i):
                ny, nx = current_borders[j]
                if abs(ny-y) < 0.01 and (x == nx + 1 or x == nx -1):
                    count -= 1

        if y % 1 == 0:
            for j in range(0, i):
                ny, nx = current_borders[j]
                if abs(nx-x) < 0.01 and (y == ny + 1 or y == ny -1):
                    count -= 1

    return count



total_score = 0
for row_i in range(ny):
    for char_i in range(nx):
        if not seen_positions[row_i][char_i]:
            current_area = []
            current_borders = []

            name = board[row_i][char_i]
            seen_positions[row_i][char_i] = 1
            current_area.append((row_i, char_i))

            discover_region(row_i, char_i, name)

            area = len(current_area)
            border = count_borders()
            total_score += area * border

print(total_score)
```
