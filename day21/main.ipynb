```python
# part 1

import os, re, sys
import numpy as np
import itertools
import math
list(itertools.permutations([1, 2, 3]))

filename = "data.txt"
with open(filename, 'r') as f:
    s = f.read()
    codes = s.split("\n")
print(codes)



coords_num = {'7': (0, 0), \
          '8': (0, 1), \
          '9': (0, 2), \
          '4': (1, 0), \
          '5': (1, 1), \
          '6': (1, 2), \
          '1': (2, 0), \
          '2': (2, 1), \
          '3': (2, 2), \
          '0': (3, 1), \
          'A': (3, 2)}

coords_dir = {
    '^': (0, 1), \
    'A': (0, 2), \
    '<': (1, 0), \
    'v': (1, 1), \
    '>': (1, 2), \
}


def solve(next, prev, coords):
    y1, x1 = coords[next]
    y0, x0 = coords[prev]
    dy, dx =  y1 - y0, x1 - x0

    result = ['^', 'v'][dy > 0] * abs(dy) + ['<', '>'][dx > 0] * abs(dx) + 'A' # volgorde maakt niet uit

    return result

def sd(all, coords):
    sol = ''
    prev = 'A'
    for char in all:
        sol += solve(char, prev, coords)
        prev = char
    print(all, " results in", sol)
    return sol


def parse_code(code):
    res = sd(sd(sd(code, coords_num), coords_dir), coords_dir)
    print(len(res))
    return res



if __name__ == "__main__":
    total = 0
    for code in codes:
        total += int(code[:3]) * len(parse_code(code))
    print(total)
```
