```python
# part 2

filename = "data.txt"
with open(filename, 'r') as f:
    s = f.read()
    codes = s.split("\n")

coords_num = {
    '7': (0, 0), \
    '8': (0, 1), \
    '9': (0, 2), \
    '4': (1, 0), \
    '5': (1, 1), \
    '6': (1, 2), \
    '1': (2, 0), \
    '2': (2, 1), \
    '3': (2, 2), \
    'X': (3, 0), \
    '0': (3, 1), \
    'A': (3, 2), \
}

coords_dir = {
    'X': (0, 0), \
    '^': (0, 1), \
    'A': (0, 2), \
    '<': (1, 0), \
    'v': (1, 1), \
    '>': (1, 2), \
}

coords_steps = {
    '^': (-1, 0), \
    'A': (0, 0), \
    '<': (0, -1), \
    'v': (1, 0), \
    '>': (0, 1), \
}


scores = {} # keep track of scores

def solve(chars, layer):
    L, prev = 0, 'A'
    for char in chars:
        L += solve_step(char, prev, layer, coords_dir)
        prev = char
    return L
    
def solve_step(next, prev, layer, coords):
    key = next + prev + str(layer)
    val = scores.get(key)
    if val:
        return val


    y1, x1 = coords[next]
    y0, x0 = coords[prev]
    dy, dx =  y1 - y0, x1 - x0
    if layer == 0:
        return 1 + abs(dy) + abs(dx) # ordering does not matter

    possibilities = [['^', 'v'][dy > 0] * abs(dy) + ['<', '>'][dx > 0] * abs(dx) + 'A',
                     ['<', '>'][dx > 0] * abs(dx) + ['^', 'v'][dy > 0] * abs(dy) + 'A']
    
    if possibilities[0] == possibilities[1]: # identical possibilities
        val = solve(possibilities[0], layer-1)
        scores[key] = val
        return val
    
    empty_spot = coords['X']
    for i in range(2):
        possibility = possibilities[i]
        y, x = y0, x0
        for char in possibility:
            dely, delx = coords_steps[char]
            y, x = y + dely, x + delx
            if (y, x) == empty_spot: # we pass the empty spot
                val = solve(possibilities[(i + 1) % 2], layer-1)
                scores[key] = val
                return val
    
    # check both possibilities
    Ls = []
    for possibility in possibilities:
        Ls.append(solve(possibility, layer-1))

    val = min(Ls)
    scores[key] = val
    return val


n_robots = 25
# for part 1: n_robots = 2
# for part 2: n_robots = 25

if __name__ == "__main__":
    total_score = 0
    for code in codes:
        code_score, prev = 0, 'A'
        for char in code:
            code_score += solve_step(char, prev, n_robots, coords_num)
            prev = char

        total_score += int(code[:3]) * code_score

    print(total_score)

# 184180
# 231309103124520
```
