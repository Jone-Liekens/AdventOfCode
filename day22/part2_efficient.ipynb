```python

# part 1

import os, re, sys
import numpy as np
import math

filename = "data.txt"
with open(filename, 'r') as f:
    s = f.read()
    
startcodes = [int(x) for x in s.split("\n")]
#startcodes = [1, 2, 3, 2024]

def step1(secr):
    return ((64 * secr) ^ secr) % 16777216

def step2(secr):
    return (math.floor(secr / 32) ^ secr) % 16777216

def step3(secr):
    return ((2048 * secr) ^ secr) % 16777216

n = 2000

prices = np.zeros((len(startcodes), n+1), dtype=int)
deltas = np.zeros((len(startcodes), n), dtype=int)


total_score = 0
for code_i in range(len(startcodes)):
    secret = startcodes[code_i]
    price = str(secret)[-1]

    prices[code_i][0] = price

    for i in range(n):
        secret = step3(step2(step1(secret)))
        price = int(str(secret)[-1])

        prices[code_i, i+1] = price
        deltas[code_i, i] = prices[code_i, i+1] - prices[code_i, i]

    total_score += secret

print(total_score)


def calculate_policy(series):
    k = len(series)
    total = 0
    for i in range(len(startcodes)):
        j = 0
        while j < n - k + 1:
            if (deltas[i, j:j+k] == series).all():
                total += prices[i, j+k]
                break
            j += 1
    return total



abcds_all = {
    (a, b, c, d): 0 \
    for a in range(-9, 10) \
    for b in range(-9, 10) \
    for c in range(-9, 10) \
    for d in range(-9, 10) \
}


for start_code_i in range(len(startcodes)):
    abcds_startcode = {}

    for i in range(0, n-3):
        sequence = tuple(deltas[start_code_i, i:i+4])

        if not abcds_startcode.get(sequence):
            abcds_startcode[sequence] = prices[start_code_i, i+4]
    

    for abcd in abcds_startcode:
        abcds_all[abcd] += abcds_startcode[abcd]



optimal_abcd = sorted(abcds_all, key=lambda x: abcds_all[x], reverse=True)

for abcd in optimal_abcd[:5]:
    print(abcd, abcds_all[abcd])
```
