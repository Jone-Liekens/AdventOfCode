```python
# part 1

import numpy as np
import re

filename = "data.txt"
with open(filename, 'r') as f:
    s = f.read()
    problems = s.split("\n\n")


token_spent = 0
for problem in problems:
    xs = [int(x[2:]) for x in re.findall(r'X\+\d+', problem)]
    ys = [int(y[2:]) for y in re.findall(r'Y\+\d+', problem)]
    goal_x = int(re.findall(r'X=\d+', problem)[0][2:])
    goal_y = int(re.findall(r'Y=\d+', problem)[0][2:])

    x1, x2 = xs
    y1, y2 = ys

    a = np.array([xs, ys])
    b = np.array([goal_x, goal_y])
    estimate = np.linalg.solve(a, b)

    na = round(estimate[0])
    nb = round(estimate[1])
    if na * x1 + nb * x2 == goal_x:
        if na * y1 + nb * y2 == goal_y:
            token_spent += 3 * na + nb


print(token_spent)
# 25357 = too low
```

```python
# part 2

import numpy as np
import re

filename = "data.txt"
with open(filename, 'r') as f:
    s = f.read()
    problems = s.split("\n\n")


token_spent = 0
for problem in problems:
    xs = [int(x[2:]) for x in re.findall(r'X\+\d+', problem)]
    ys = [int(y[2:]) for y in re.findall(r'Y\+\d+', problem)]
    goal_x = int(re.findall(r'X=\d+', problem)[0][2:])
    goal_y = int(re.findall(r'Y=\d+', problem)[0][2:])

    x1, x2 = xs
    y1, y2 = ys


    nal = y2 - x2
    nbl = x1 - y1

    if nal * xs[0] + nbl * xs[1] < 0:
        nal, nbl = -nal, -nbl
    step = nal * xs[0] + nbl * xs[1]

    # so that:
    # [x1 x2] * [nal]  = [step]
    # [y1 y2]   [nbl]    [step]

    large = 10000000000000
    extra = large % step
    n_extra = large // step

    a = np.array([xs, ys])
    b = np.array([goal_x, goal_y]) + extra
    estimate = np.linalg.solve(a, b)

    na = round(estimate[0])
    nb = round(estimate[1])
    if na * xs[0] + nb * xs[1] == goal_x + extra:
        if na * ys[0] + nb * ys[1] == goal_y + extra:
            token_spent += 3 * (na + nal * n_extra) + (nb + nbl * n_extra)

print(token_spent)
```
