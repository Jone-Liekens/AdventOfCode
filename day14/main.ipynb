```python
# part 1

import re
import numpy as np

filename = "data.txt"
with open(filename, 'r') as f:
    s = f.read()
    robots = s.split("\n")


nx, ny = 101, 103
if filename == "test.txt":
    nx, ny = 11, 7


quadrant_counts = np.zeros((4), dtype=int)

for robot in robots:
    position, velocity = robot.split(" ")
    x = [int(x[2:]) for x in re.findall(r'p=-?\d+', position)][0]
    y = [int(y[1:]) for y in re.findall(r',-?\d+', position)][0]

    vx = [int(x[2:]) for x in re.findall(r'v=-?\d+', velocity)][0]
    vy = [int(y[1:]) for y in re.findall(r',-?\d+', velocity)][0]

    # position after 100 steps
    x = (x + 100 * vx) % nx
    y = (y + 100 * vy) % ny

    if x < nx // 2:
        if y < ny // 2:
            quadrant_counts[0] += 1
        elif y > ny // 2:
            quadrant_counts[1] += 1
    elif x > nx // 2:
        if y < ny // 2:
            quadrant_counts[2] += 1
        elif y > ny // 2:
            quadrant_counts[3] += 1

print(quadrant_counts)
print(np.prod(quadrant_counts))

```
