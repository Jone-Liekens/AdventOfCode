```python
filename = 'data.txt'
with open(filename, 'r') as f:
    s = f.read()
    board = s.split("\n")

nx = len(board[0])
ny = len(board)
print(nx, ny)


new_obstacles = [[' ' for i in range(nx)] for j in range(ny)]

# search start position
for yi in range(ny):
    for xi in range(nx):
        if board[yi][xi] == '^':
            x, y = xi, yi
            start_x, start_y = x, y

# directions (0 = up, 1, = right, 2 = down, 3 = left)
# and their correspondence to steps in x and y (y direction is counterintuitive)
dir = 0
dirs = [(0, -1), (1, 0), (0, 1), (-1, 0)]
dirnames = ['u', 'r', 'd', 'l']


def print_simple_board(board):
    print(" " + "_" * nx + " ")
    for row in board:
        print('|', end='')
        for item in row:
            print(item, end='')
        print('|')
    print(" " + "_" * nx + " \n")

def print_complex_board(board):
    for row in board:
        for item in row:
            print(item + ' ' * (4 - len(item)), end='|') 
        print("\n" + "-----" * len(row))
    print("\n\n")

print_simple_board(board)
```

```python

def check_loop(x, y, dir):
    # do we get stuck in a loop if we put an obstacle right in front of us?

    # if we use small test data, we print extra debug info
    debug = filename == 'test.txt'
     
    
    board_copy =          [['' for i in range(nx)] for j in range(ny)]
    if debug: test_path = [[' ' for i in range(nx)] for j in range(ny)]

    # where is "right in front of us"?
    dx, dy = dirs[dir]
    new_x, new_y = x + dx, y + dy

    # if "right in front us" is outside the board, it is not possible to put an obstacle
    if new_x < 0 or new_y < 0 or new_x >= nx or new_y >= ny:
        return False
    else:
        board_copy[new_y][new_x] = '####'
        if debug: test_path[new_y][new_x] = '!' 
    
    for yi in range(ny):
        for xi in range(nx):
            if board[yi][xi] == '#':
                board_copy[yi][xi] = "####"
                if debug: test_path[yi][xi] = '#' 



    dir = 0
    x, y = start_x, start_y
    if debug: test_path[start_y][start_x] = "^"
    while True:
        
        dx, dy = dirs[dir]
        new_x, new_y = x + dx, y + dy

        if dirnames[dir] in board_copy[y][x]:
            if debug:
                test_path[y][x] = "o"
                print_simple_board(test_path)
                
            return True
        else:
            board_copy[y][x] += dirnames[dir]

        if new_x < 0 or new_y < 0 or new_x >= nx or new_y >= ny:
            if debug: print_simple_board(test_path)
            return False

        if board_copy[new_y][new_x] != '####':
            x, y = new_x, new_y
            if debug: test_path[y][x] = 'x'
        elif board_copy[new_y][new_x] == '####':
            dir = (dir + 1) % 4
        else:
            import sys; sys.exit()



count_new_obstacles_added = 0
while True:
    dx, dy = dirs[dir]
    new_x, new_y = x + dx, y + dy

    if check_loop(x, y, dir):
        new_obstacles[new_y][new_x] = 'O'
        count_new_obstacles_added += 1

    if new_x < 0 or new_y < 0 or new_x >= nx or new_y >= ny:
        break

    if board[new_y][new_x] != '#':
        x, y = new_x, new_y
    elif board[new_y][new_x] == '#':
        dir = (dir + 1) % 4


print_simple_board(new_obstacles)
print(count_new_obstacles_added)
```

```python
# some locations (x, y) will have been checked twice
# here we check how many new obstacles are unique,
# by counting the new obstacles on the map
# instead of how many times we added an obstacle

count_unique_obstacles = 0
for y in range(ny):
    for x in range(nx):
        if new_obstacles[x][y] == "O":
            count_unique_obstacles += 1
print(count_unique_obstacles)
```
