```python
# part 1

import os, re, sys
import numpy as np
import itertools
import math
list(itertools.permutations([1, 2, 3]))

filename = "test.txt"
with open(filename, 'r') as f:
    s = f.read()
    codes = s.split("\n")
print(codes)



coords_num = {'7': (0, 0), \
          '8': (0, 1), \
          '9': (0, 2), \
          '4': (1, 0), \
          '5': (1, 1), \
          '6': (1, 2), \
          '1': (2, 0), \
          '2': (2, 1), \
          '3': (2, 2), \
          '0': (3, 1), \
          'A': (3, 2)}

coords_dir = {
    '^': (0, 1), \
    'A': (0, 2), \
    '<': (1, 0), \
    'v': (1, 1), \
    '>': (1, 2), \
}

coords_steps = {
    '^': (-1, 0), \
    'A': (0, 0), \
    '<': (0, -1), \
    'v': (1, 0), \
    '>': (0, 1), \
}
```

```python


def solve_numeric(next, inputprev, n_layers):
    y1, x1 = coords_num[next]
    y0, x0 = coords_num[inputprev]
    dy, dx =  y1 - y0, x1 - x0

    possibilities = [['^', 'v'][dy > 0] * abs(dy) + ['<', '>'][dx > 0] * abs(dx) + 'A',
                     ['<', '>'][dx > 0] * abs(dx) + ['^', 'v'][dy > 0] * abs(dy) + 'A'] # volgorde maakt uit
    
    if possibilities[0] == possibilities[1]:
        #print("only one option: ", possibilities[i+1])

        return solve_range(possibilities[0], n_layers) # twee zijn identiek
    
    # check if we pass an empty space
    for i in range(2):
        possibility = possibilities[i]
        y, x = y0, x0
        for char in possibility:
            dely, delx = coords_steps[char]
            y, x = y + dely, x + delx
            if (y, x) == (3, 0):
                #print("only one option: ", possibilities[i+1])

                return solve_range(possibilities[(i + 1) % 2], n_layers)
    
    Ls = []
    for possibility in possibilities:
        Ls.append(solve_range(possibility, n_layers))

    print("in 1, looking for ", next, "from ", inputprev, " gives us ", possibilities, "with values ", Ls)
    return min(Ls)


def solve_range(chars, layer):
    L, prev = 0, 'A'
    for char in chars:
        L += solve(char, prev, layer)
        prev = char
    return L
    

def solve(next, prev, layer = 25):
    y1, x1 = coords_dir[next]
    y0, x0 = coords_dir[prev]
    dy, dx =  y1 - y0, x1 - x0
    if layer == 0:
        possibility = ['^', 'v'][dy > 0] * abs(dy) + ['<', '>'][dx > 0] * abs(dx) + 'A' # volgorde maakt niet uit
        return len(possibility)

    possibilities = [['^', 'v'][dy > 0] * abs(dy) + ['<', '>'][dx > 0] * abs(dx) + 'A',
                     ['<', '>'][dx > 0] * abs(dx) + ['^', 'v'][dy > 0] * abs(dy) + 'A']# volgorde maakt uit
    
    if possibilities[0] == possibilities[1]:
        return solve_range(possibilities[0], layer-1) # twee zijn identiek
    
    # check if we pass an empty space
    for i in range(2):
        possibility = possibilities[i]
        y, x = y0, x0
        for char in possibility:
            dely, delx = coords_steps[char]
            y, x = y + dely, x + delx
            if (y, x) == (0, 0):

                return solve_range(possibilities[(i + 1) % 2], layer-1)
    
    # allright, seems like we really should check both options
    Ls = []
    for possibility in possibilities:
        Ls.append(solve_range(possibility, layer-1))

    return min(Ls)



if __name__ == "__main__":
    total_score = 0
    for code in codes:
        prev = 'A'
        total = 0
        for char in code:
            total += solve_numeric(char, prev, 1)
            prev = char

        total_score += int(code[:3]) * total
        print("--------------------------------------------------------------------")
        print("TOTAL", total)
        print("--------------------------------------------------------------------")
    print(total_score)
```
