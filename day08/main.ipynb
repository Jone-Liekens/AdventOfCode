```python
import numpy as np

filename = "data.txt"
with open(filename, 'r') as f:
    s = f.read()
    rows = s.split("\n")
print(rows)


n_rows = len(rows)
n_cols = len(rows[0])


freq_pos = dict()
for row_i in range(n_rows):
    for elem_i in range(n_cols):
        freq = rows[row_i][elem_i]
        if freq != '.':
            if freq_pos.get(freq) is None:
                freq_pos[freq] = [(row_i, elem_i)]
            else:
                freq_pos[freq].append((row_i, elem_i))


print(freq_pos)





antinodes = np.zeros((n_rows, n_cols))


def check_in_board(pos):
    if pos[0] >= 0 and pos[0] < n_rows and pos[1] >= 0 and pos[1] < n_cols:
        antinodes[pos[0], pos[1]] = 1
        return True
    return False

for freq in freq_pos:
    positions = freq_pos[freq]
    n_pos = len(positions)
    for i in range(n_pos):
        for j in range(i+1, n_pos):
            pos1 = positions[i]
            pos2 = positions[j]

            delta_0 = pos1[0] - pos2[0]
            delta_1 = pos1[1] - pos2[1]

            n = 0
            in_board = True
            while in_board:
                pos = (pos1[0] + n * delta_0, pos1[1] + n * delta_1)
                in_board = check_in_board(pos)
                n = n - 1

            n = 0
            in_board = True
            while in_board:
                pos = (pos1[0] + n * delta_0, pos1[1] + n * delta_1)
                in_board = check_in_board(pos)
                n = n + 1



print(antinodes)
print(sum(sum(antinodes)))


```
